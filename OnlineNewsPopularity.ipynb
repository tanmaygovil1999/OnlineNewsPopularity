```python
from scipy import stats
import numpy as np
```

```python
def remove_outliers(df):
  for col in df.select_dtypes(include='number'):
    df[col] = np.where(np.abs(stats.zscore(df[col])) < 3 , df[col], df[col].mean())
    return df
```

### Part-1

```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
import spacy
from gensim import corpora, models
import string
```

```python
nltk.downloader.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')
sia = SentimentIntensityAnalyzer()
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
df["url_heading"] = df["url"].str.split('/', n=7, expand=True)[6]
```

```python
df["clean_heading"] = df["url_heading"].str.replace("-", " ")
```

```python
sentimental_scores = []
topics = []
ner_counts = []
```

```python
def entity_extraction(text):
  doc = nlp(text)
  # entities = [ent.label_ for ent in doc.ents]
  return len(doc.ents)
```

```python
def clean_heading(heading):
  heading = heading.lower()
  clean_heading = heading.replace("-", " ")
  return clean_heading
```

```python
def calculate_sentimental_score(text):
  return sia.polarity_scores(text)
```

```python
def perform_topic_modelling(text):
  tokens = word_tokenize(text)
  dictionary = corpora.Dictionary([tokens])
  corpus = [dictionary.doc2bow(tokens)]
  lda_model = models.LdaModel(corpus, num_topics=5 , id2word = dictionary, passes=20)
  topics = lda_model.get_document_topics(dictionary.doc2bow(tokens))
  topic_proportion = [topic[1] for topic in topics]
  # print(topic_proportion)
  # topics = lda_model.print_topics(num_words=5)
  return topic_proportion
```

```python
def ner_count(text):
  return entity_extraction(text)
```

```python
perform_topic_modelling("twitter ben frankl birthday")
```

```python
df["compound"], df['pos'], df['neg'], df['neu'] = zip(*df["clean_heading"].apply(calculate_sentimental_score))
df["Topics"] = df["clean_heading"].apply(perform_topic_modelling)
df["NER_Count"] = df["clean_heading"].apply(ner_count)
```

```python
from transformers import BertModel
import torch
from transformers import AutoTokenizer
```

```python
model = BertModel.from_pretrained('bert-base-uncased')
tokenizer = AutoTokenizer.from_pretrained('bert-base-cased')
```

```python
def bert_embeddings(text):
  input = tokenizer(text, return_tensors='pt', max_length = 512, truncation = True)
  with torch.no_grad():
    output = model(**input)
    embeddings = output.pooler_output.detach().numpy().squeeze()
  return embeddings
```

```python
df["bert_embedding"] = df["clean_heading"].apply(bert_embeddings)
```

```python
from scipy.stats import pearsonr
```

```python
correlations = {}
popularity_measure = "shares"

for feature in ['compound', "pos", "neg", "neu", "Topics", "NER_Count"]:
  correlation, _ = pearsonr(df[feature], df[popularity_measure])
  correlations[feature] = correlation

strongest_correlation = sorted(correlations.items(), key = lambda x: abs(x[1]), reverse=True)
print(strongest_correlation)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error , r2_score
```

```python
extra_features = ["url_heading", "shares"]
total_columns = [item for item in list(df.columns) if item not in extra_features]
```

```python
X = df[total_columns]
y = df["shares"]

X_train, X_test, y_train, y_test = train_test_split(X, y)
models = {"LinearRegression": LinearRegression(),
    "RandomForestRegressor":  RandomForestRegressor(),
          "MLPRegressor": MLPRegressor()
          }
for name , model in models.items():
  model.fit(X_train, y_train)
  preds = model.predict(X_test)
  mse = mean_squared_error(y_test, preds)
  r2 = r2_score(y_test, preds)
```
